```python
#This model is rescricted to a closed economy without government (taxes and public expenditure)
```

```python
#importing librarys
import matplotlib.pyplot as plt
import numpy as np
```

```python
#models parameters

#these are the only constants that is mutable in the model.
#fraction of income to pay the capital
a = 0.33
#depreciation
d = 0.15
#savings 
s = 0.3 

#population growth
n = 0.05
#this parameters are tuned to graph visualization, in the reality, s is hardly above 0.15% and so on with depreciation
```

```python
#producions function
def production_funcion(L,K):
    Y = (L**(1-a))*K**a
    return Y
```

```python
def intensive_function(k):
    y= k**a
    return y

def expenditure(k):
    c = intensive_function(k)*(1-s)
    return c
```

```python
#here is a function to discover the optimun capital per worker of the model, the steady state k and steady state y
variable = 0
while True:
    t = (round(s*intensive_function(variable),5) - round((n+d)*variable,5)) == 0
    if t == True and variable != 0:
        k_ss = variable
        break
    else:
        variable = variable + 0.000001

print('the capital of steady state is: ',round(k_ss,2))
y_ss = intensive_function(k_ss)
print('the yield of steady state is: ',round(y_ss,2))

#this function retrieves the x axis in the graph to mantein a good visualization, despite the parameters above. K/L is endogenous and so is Y/L

#the exogenous parameters in the model are A (which i did not considered in this implementation because it would just be another constant to tune.
#a better version of the production function would be Y = AL^(1-a)K^a
#and the intensive function y = Ak^a
#A can be endogenous in a more complex implementation, beeing A = E x T, T being technology(P&D model) and E being eficiency

k_values= np.arange(0,k_ss + (1/3)*k_ss,0.001)
```

```python
#fundamental equation of solows model delta k = s*y-(d+n)*k

#like the omission of A, here there is a omission of another variable g = (dA/dT)/A, the growth of human capital.
#A complete version would look like k = s*y-(d+n+g)*k

y_values = [intensive_function(k) for k in k_values]
sy_values = [s*intensive_function(k) for k in k_values]
dk = [(n+d)*k for k in k_values]
```

```python
plt.plot(k_values,sy_values, 'b')
plt.plot(k_values,dk, 'tab:orange')
plt.plot(k_values,y_values, 'g')
plt.xlabel('Capital per worker')
plt.ylabel('Yield per worker')
plt.title('graph of capital per worker and yield per worker in the solows model')
plt.grid(True)
plt.plot(k_ss,y_ss, marker='o')
plt.plot(k_ss,0, marker='o')
plt.plot(0,y_ss, marker = 'o')
plt.plot([k_ss, k_ss], [y_ss, 0], 'k--')

plt.plot([k_ss,0],[y_ss,y_ss], 'k--')
plt.axvline(x=0, color='k')  # Linha vertical em x = 0, cor vermelha e estilo pontilhado
plt.axhline(y=0, color='k')

plt.text(k_ss+0.01*k_ss,1/100*k_ss,f'capital in steady state, equals {round(k_ss,2)}')
plt.text(0,y_ss+y_ss/100*y_ss, f'yield in steady state, equals {round(y_ss,2)}')
plt.show
```

```python
#the orange line is the depreciation of the capital, the orange line is capital formation.
#in the equilibrium of the model, there is no change in the capital per worker, which remains constant.
#the green line reflects the yield per worker in equilibrium of this model. 
```

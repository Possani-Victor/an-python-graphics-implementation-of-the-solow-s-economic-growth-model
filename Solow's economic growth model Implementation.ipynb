```python
#This model is rescricted to a closed economy without government (taxes and public expenditure)
```

```python
#importing librarys
import matplotlib.pyplot as plt
import numpy as np
```

```python
#models parameters

#these are the only constants that is mutable in the model.
#fraction of income to pay the capital
a = 0.33
#depreciation
d = 0.4
#savings 
s = 0.5 

#population growth
n = 0.05
#value of capital per worker, this exists to be the x axis in the model, so, if the constants not converge, try modifieing this one.
k_values = np.arange(0,5,0.001)

#this parameters are tuned to graph visualization, in the reality, s is hardly above 0.15% and so on with depreciation
```

```python
#producions function
def production_funcion(L,K):
    Y = (L**(1-a))*K**a
    return Y
```

```python
def intensive_function(k):
    y= k**a
    return y

def expenditure(k):
    c = intensive_function(k)*(1-s)
    return c
```

```python
#here is a function to discover the optimun capital per worker of the model, the steady state k and steady state y
for k in k_values:
    tolerance = 0.000000001 #this tolerance existins because k is not a continuous range, instead is a discrete list of constants ranging from a to b
    t = (s*intensive_function(k) - (n+d)*k) < tolerance 
    if t == True and k != 0:
        k_ss = k
        break

print('the capital of steady state is: ',round(k_ss,2))
y_ss = intensive_function(k_ss)
print('the yield of steady state is: ',round(y_ss,2))
```

```python
#fundamental equation of solows model delta k = s*y-(d+n)*k
y_values = [intensive_function(k) for k in k_values]
sy_values = [s*intensive_function(k) for k in k_values]
dk = [(n+d)*k for k in k_values]
```

```python
plt.plot(k_values,sy_values, 'b')
plt.plot(k_values,dk, 'tab:orange')
plt.plot(k_values,y_values, 'g')
plt.xlabel('Capital per worker')
plt.ylabel('Yield per worker')
plt.title('graph of capital per worker and yield per worker in the solows model')
plt.grid(True)
plt.plot(k_ss,y_ss, marker='o')
plt.plot(k_ss,0, marker='o')
plt.plot(0,y_ss, marker = 'o')
plt.plot([k_ss, k_ss], [y_ss, 0], 'k--')

plt.plot([k_ss,0],[y_ss,y_ss], 'k--')
plt.axvline(x=0, color='k')  # Linha vertical em x = 0, cor vermelha e estilo pontilhado
plt.axhline(y=0, color='k')

plt.text(k_ss+0.1,0.05,f'capital in steady state, equals {round(k_ss,2)}')
plt.text(1,y_ss + 0.2, f'yield in steady state, equals {round(y_ss,2)}')
plt.show
```
